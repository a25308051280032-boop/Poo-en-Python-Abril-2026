# Lenguaje Python, crear la clase "animal" con los atributos "nombre" , "raza" crear el conductor animal, las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
class Animal:
    """Clase que representa a un animal genérico."""

    # El constructor (__init__) inicializa los atributos al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} (raza {self.raza}) dice: ¡Guau, guau! 🐱‍🏍")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre} está corriendo muy rápido. 🐱‍👓")

    def Morder(self):
        print(f"{self.nombre} esta mordiendo a al vecina 🤦‍♀️")

# Creación de los objetos (instancias de la clase Animal)
pitbull = Animal("Roco", "Pitbull")
chihuahua = Animal("Chispita", "Chihuahua")

#creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "chafa")

# Uso de las funciones con los objetos creados
pitbull.ladrar()
pitbull.correr()

print("-" * 20) # Línea divisoria para organizar la salida

chihuahua.ladrar()
chihuahua.correr()
perrocallejero.Morder()
print("Programa ralizado por Iris Chavez NL = 11")
```
